```python
local_crs='EPSG:2154'
country_code ='FR'
```

```python
data_dir_in, data_dir_out = set_up_directories('FR','Auxerre')
```

```python
city_df = get_city_selection_df(target_city_name=['Auxerre'])
city_df
```

```python
city = get_city(get_specific_place_name(city_df),local_crs)
city_subset = get_city(['Auxerre,France','Monéteau,France','Perrigny,France','Saint-Georges-sur-Baulche,France','Appoigny,France'],local_crs)
```

```python
city.plot()
```

```python
add_Quicksand_font()

not_city = get_city_negative(city)

fig, ax = plt.subplots(1, 1, figsize=(10,10))
_ = ax.axis('off')

# reset zoom level (due to not_city box)
xlim, ylim = get_map_limits(city)
ax.set_xlim(xlim)
ax.set_ylim(ylim)

not_city.plot(
    ax=ax,
    color='white',
    edgecolor='none',
    alpha=.4,
)

city.plot(
    ax=ax,
    color='none',
    edgecolor='black',
    linewidth=.4,
    alpha=1
)
city_subset.plot(
    ax=ax,
    color='green',
    edgecolor='black',
    linewidth=0,
    alpha=.4
)

# LEGEND
## legend elements
hexagon1 = Line2D([0], [0], color='red', marker='h', linewidth=0, markersize=11, alpha=0.8)
point = Line2D([0], [0], color='w', marker='o',markeredgecolor='k', linewidth=1, markersize=8, alpha=0.8)
line = Line2D([0], [0], color='black', linewidth=.6, markersize=1, alpha=1 ,label='a')
cmap_gradient = get_gradient_cmap()

## labels
if country_code == 'FR':
    labels = ["Potentiel emplacement station","Station existante",'Score de demande latente', 'Limite commune'] 
else:
    labels = ['Bikeshare station', 'Bikeshare potential', 'City limit']
handles = [hexagon1, point,cmap_gradient,line]

# add_legend(ax, labels, handles,title='Saint-Étienne Métropole', location='lower right')
add_city_names(city,font_size=7)


add_north_arrow(ax)
add_scalebar(ax)

# basemap 
cx.add_basemap(
    ax,
    crs=city.crs,
    # source = cx.providers.OpenStreetMap.Mapnik, # #cx.providers.Stamen.TonerLite, #cx.providers.Stamen.Toner
    source = cx.providers.Esri.WorldGrayCanvas,
    attribution=False,
    attribution_size=6
)

plt.savefig(os.path.join(data_dir_out,f"Auxerre_VLS_subset_v1.png"),dpi=300,facecolor='w', bbox_inches='tight',)
```
